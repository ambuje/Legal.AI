#Final Annotations 'BIO' csv

```
import pandas as pd
```

```
df=pd.read_json('/content/hk_45-final_annotated_contracts.json1', lines = True)
```

```
df
```

```
words = []
tags = []
```

```
for i in range(len(df)):
  text = df['text'][i].strip()
  #text = ' between eden hospital health services corporation and eden medical center '.strip()
  #print(text)
  textl = list(text)
  #print(textl)
  tagl = ['O'] * len(textl)
  lb = df['labels'][i]
  
  for k in range(len(lb)):
    t = lb[k]
    sp = False
    tagl[t[0]] = 'B-'+t[2]
    for l in range(t[0]+1,t[1]):
      if sp:
        tagl[l] = 'I-'+t[2]
      else:
        tagl[l] = 'B-'+t[2]
        if textl[l] == ' ': sp = True
  #print(tagl)

  word = ''
  tag = ''
  for j in range(len(textl)):
    #if tagl[j] == 
    if textl[j] == ' ':
      words.append(word)
      word = ''
      tags.append(tag)
    else:
      word = word + textl[j]
      tag = tagl[j]
    
  words.append(word)
  tags.append(tag)
  words.append('')
  tags.append('')
  #print(words)
  #print(tags)
  print(i)
```

```
df1 = pd.DataFrame(list(zip(words, tags)))
```
